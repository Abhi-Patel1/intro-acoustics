# Generate sine wave

A basic overview of creating sine wave signals in MATLAB

## Theory

Sine waves are incredibly useful in acoustics because any sound can be constructed from the superposition of sine waves. Sine and cosine waves are constructed using the following equations:  

$y_{sin} = a*\sin(2*\pi*f*t)$  
$y_{cos} = a*\cos(2*\pi*f*t)$  

Where:  
$a$ = signal amplitude ($\mu$Pa)  
$f$ = signal frequency (Hz)  
$t$ = time vector (s)  

## Application

### Inputs

Here's an example of how to construct sine waves in MATLAB. We have to begin by defining some very basic qualities of our simulated signal.

```matlab
% sampling frequency
Fs = 100;

% sampling interval
dt = 1/Fs; 

% signal duration
d = 1; 

% sampling timesteps
t = 0:dt:d-dt;

% signal frequency
f = 1; 

% signal amplitude
a = 1; 
```

### Generate waves

Apply the above formula to general both sine and cosine waves

```matlab
% Generate the sine wave
y_sin = a*sin(2*pi*f*t);

% Compare with a cosine wave
y_cos = a*cos(2*pi*f*t);
```

### Plot waves

```matlab
plot(t, y_sin);
ylim([-2 2])
ylabel('Amplitude')
xlabel('Time')
hold on
plot(t,y_cos)
line([0 2], [0,0],'Color','black')
legend('sine', 'cosine')
grid on
```
